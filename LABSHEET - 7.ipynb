PLOT CATEGORIES IN SEABORN
1. RELATIONAL PLOTS : This is used to understand the realation between two variables
2. CATEGORIACAL PLOTS : This deals woth categorical variables and how they can be visualized
3. DISTRIBUTIKON PLOTS : This plot is udes for examininf univariate and bivarirate distributions
4. MATRIX PLOTS : It is an array of scatterplots
5. REGRESSION PLOTS : Primarily intended to add a visual guide and helps to emphasise patterns.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
%matplotlib inline
```

SIMPLE PLOTTING USING SEABORN

```python
dates = ['1981-01-01','1981-01-02','1981-01-03','1981-01-04','1981-01-05','1981-01-06','1981-01-07','1981-01-08','1981-01-09','1981-01-10']
min_temp = [20.7,17.9,18.8,14.6,15.8,15.8,15.8,17.4,21.8,20.0]
max_temp = [34.7,28.9,31.8,25.6,28.8,21.8,22.8,28.4,30.8,32.0]
fig,axes = plt.subplots(nrows=1,ncols=1,figsize=(15,10))
axes.plot(dates,min_temp,label = 'Min Temperature',marker = '*')
axes.plot(dates,max_temp,label = 'Max Temperature',marker = 's')
axes.legend()
```

RELATIONAL PLOTS
1. LINE PLOT : It is one of the most basix plot in seaborn library, used to plot continuos data, visualises data in form of time series

```python
sns.set(style="dark")
fix,axes = plt.subplots(ncols=2,nrows=1,figsize=(15,10))
#LOADING DATA WITH SEABORN
df = sns.load_dataset("tips")
print(df.head())
#LINEPLOT
sns.lineplot(x="total_bill",y="tip",hue="size",style="time",data=df,ax=axes[0]).set_title("Line Plot")
#SCATTER PLOT
Sct_plt = sns.scatterplot(x="total_bill",y="tip",hue="size",style="time",data=df,ax=axes[1]).set_title("Scatter Plot")
#SAVING PLOT
Sct_plt.figure.savefig('Scatter_plot.png')
print('Plot saved')
```

CATEGORICAL PLOTS
USED FOR VISUALIZING THE REALATIONSHIP BETWEEN VARIABLES, VARIABLES CAN BE CATEGORICAL,NUMERICAL OR GROUP OR CLASS OR DIVISION

```python
sns.set_style('darkgrid')
fig,ax = plt.subplots(nrows=5,ncols=2)
fig.set_size_inches(18.5,10.5)
# Tips dataset contains information about the people who probably had food
df = sns.load_dataset('tips')
# barplot - basically used to aggregate the categorical and ret
sns.barplot(x='sex',y='total_bill',data=df,palette='plasma',estimator = np.std,ax=ax[0,0]).set_title('Bar Plot')
# countplot - Counts the categories and returns a count of their occurences
sns.countplot(x='sex',data=df,ax=ax[0,1]).set_title('Count Plot')
# boxplot - known as thje box and whisker plot shows the distribution of the quantitative data 
sns.boxplot(x='day',y='total_bill',data=df,hue='smoker',ax=ax[1,0]).set_title('Box Plot')
# Simiplar to boxplot except that it provides a 
sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True,ax=ax[1,1]).set_title('Violin Plot')
# Stripplot - Scatter plot based on category
sns.stripplot(x='day',y='total_bill',data=df,jitter=True,hue='smoker',dodge=True,ax=ax[2,0]).set_title('Strip Plot')
# Swarmplot - Similar to stripplot except the fact that the points are adjusted
sns.swarmplot(x='day',y='total_bill',data=df,ax=ax[2,1]).set_title('Swarm Plot')
# Combining the idea of violin plot and stripplot to form this plot
sns.violinplot(x='day',y='total_bill',data=df,ax=ax[3,0])
sns.swarmplot(x='day',y='total_bill',data=df,color='black',ax=ax[3,1]).set_title('Combined Plot')
# Density plot
sns.boxenplot(x='day',y='total_bill',color="b",scale='linear',data=df,ax=ax[4,0])
# Ridgeplot
sns.pointplot(x='day',y='total_bill',color="b",hue="sex",data=df,ax=ax[4,1])
# Catplot - Provies a parameter called 'kind' to choose the kind of plot
sns.catplot(x='day',y='total_bill',data=df,kind='bar')
Sct_plt.figure.savefig('ALL DAV PLOTS.png')
print('Plots Saved')
```

DISTRIBUTION PLOTS

JOINT,DIS,PAIR,RUG PLOTS

```python
sns.set_style('whitegrid')
df = sns.load_dataset('iris')
print(df.head())
sns.distplot(df['petal_length'],kde=True,color='red',bins=30).set_title('Dist Plot')
jointgrid = sns.JointGrid(x='petal_length',y='petal_width',data=df)
jointgrid.plot_joint(sns.scatterplot)
jointgrid.plot_marginals(sns.displot)
g = sns.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
g.fig.suptitle("Joint Plot")
g = sns.pairplot(df,hue="species",palette='coolwarm')
g.fig.suptitle("Pair Plot - 1")
g.add_legend()
pairgrid = sns.PairGrid(data=df)
pairgrid = pairgrid.map_offdiag(sns.scatterplot)
pairgrid = pairgrid.map_diag(plt.hist)
pairgrid = sns.PairGrid(data=df)
pairgrid = pairgrid.map_upper(sns.scatterplot)
pairgrip = pairgrid.map_diag(plt.hist)
pairgrid = pairgrid.map_lower(sns.kdeplot)
g=sns.PairGrid(data=df)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
sns.rugplot(df['petal_length'])
plt.show()
```

MATRIX PLOTS

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(15,10))
df1 = sns.load_dataset('flights')
df2 = sns.load_dataset('iris')
df11 = pd.pivot_table(values='passengers',index='month',columns='year',data=df1)
dfc1 = df1.corr
dfc2 = df2.corr
sns.heatmap(df11,cmap='YlGnBu',linecolor='r',linewidths=0.5,annot=True,fmt='d',square=True,ax=ax[0,0]).set_title('Heat Map Flights')
sns.heatmap(dfc2,linecolor='black',linewidths=1.0,annot=True,ax=ax[0,1]).set_title('Heat Map Iris')
mask1 = np.triu(dfc2)
sns.heatmap(dfc2,annot=True,mask=mask1,ax=ax[1,0],cmap='coolwarm').set_title('Heat Map Lower Triangle')
mask2 = np.tril(dfc2)
sns.heatmap(dfc2,annot=True,mask=mask2,ax=ax[1,1],cmap='Y1GnBu').set_title('Heat Map Upper Triangle')
sns.clustermap(df11,cmap='RdY1Gn')
sns.clustermap(df11,cmap='plasma',standard_scale=1)
```
