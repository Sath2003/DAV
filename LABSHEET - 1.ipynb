EXERCISE - 1 : CREATE A DATA FRAME FOR MONTH AND TOTAL PROFIT. READ TOTAL PROFIT OF ALL THE MONTHS SHOW IT USING A LINE PLOT.TOTAL PROFIT DATA PROVIDED FOR EACH MONTH.
X - MONTH NUMBER
Y - TOTAL PROFIT


```python
import matplotlib.pyplot as plt
import numpy as np
Month = np.array(["Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec"])
Profit = np.array([200,100,50,800,10000,2000,14000,-1000,1800,2000,2200,24000])
plt.plot(Month,Profit,marker = "*",ms = 20,mfc = 'm')
plt.title("Month vs Profit")
    
```


EXERCISE - 2 : GET TOTAL PROFIT OF ALL MONTHS AND SHOW LINE PLOT WITH FOLLOWING STYLE
1. LINE - DOTTED, RED
2. SHOW LEGEND AT LOWER RIGHT LOCATION
3. x - MONTH NUMBER
4. y - SOLD UNITS NUMBERS
5. ADD CIRCLE MARKER
6. LINE MARKER RED IN COLOR

```python
Month = np.array(["Jan","Feb","Mar","Apr","May","June","July","Aug","Sept","Oct","Nov","Dec"])
Profit = np.array([200,100,50,800,10000,2000,14000,000,1800,2000,2200,24000])
plt.plot(Month,Profit,'o-.r')
plt.title("Month vs Profit")
plt.legend(["Graph"],loc = "lower right")
```

r -red, b-blue,w-white,m-magenta,k-black,y-yellow,c-cyan,g-green
