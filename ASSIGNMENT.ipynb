1. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016. Sample Financial data


```python
data = pd.read_csv('/workspaces/DAV/fdata.csv')
data.head()
```

```python
dates = data['Date']
opens = data['Open']
high = data['High']
low = data['Low']
close = data['Close']
plt.plot(dates,opens,label="Open",marker='o',)
plt.plot(dates,close,label="Close",marker='*')
plt.plot(dates,high,label="High",marker='^')
plt.plot(dates,low,label="Low",marker='s')
plt.xlabel("Dates")
plt.ylabel("Profit or Loss")
plt.title("Financial Data")
plt.legend()
plt.show()
```

3. Write a Python program to create a stacked bar plot with error bars.Note: Use bottom to stack the womens bars on top of the mens bars.

```python
data3 = pd.read_csv('/workspaces/DAV/census.csv')
num=[1,2,3,4,5]
Means_M=[22,30,35,35,26]
Means_W=[25,32,30,35,29]
Men_Standard_Deviation=[4,3,4,1,5]
Women_Standard_Deviation=[3,5,2,3,3]

plt.bar(num,Means_M,color="orange")
plt.bar(num,Means_W,color="black",bottom=Means_M)
plt.bar(num,Men_Standard_Deviation,color="blue")
plt.bar(num,Women_Standard_Deviation,color="red",bottom=Men_Standard_Deviation)
plt.show()
```
