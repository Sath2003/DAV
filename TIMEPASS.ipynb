```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/workspaces/DAV/RHS_2015_ch-1_table_2a.csv')
data.head()
```

```python
width = 1
state = data['India / State/ Union Territory']
pop = data['Population - 2011 in Million']
plt.barh(state,pop,label = "Population",color = 'orange')
plt.title("States v/s Population")
plt.xlabel("States")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
rate = data['Decadal Population Growth Rate - 2001-2011']
plt.barh(state,rate,label = "Population Rate")
plt.title("States v/s Population Rate")
plt.xlabel("States")
plt.ylabel("Population")
plt.legend()
```

```python
density = data['Population Density (per sq.km) - 2011']
plt.barh(state,density,label = "Density",color = 'r')
plt.title("States v/s Population Density")
plt.xlabel("States")
plt.ylabel("Density")
plt.legend()
```
