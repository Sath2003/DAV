PIE CHARTS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

RECOMMENDED TO USE WHEN WE HAVE 5 OR LESS VALUES TO PLOT

1. Plotting the pie chart graph 

```python
slice=[30,40,20,50]
plt.pie(slice)
plt.show()
```

2. ADDING LABELS TO PIE CHART

```python
labels = ['Thirty','Fourty','Twenty','Fifty']
plt.pie(slice, labels = labels)
plt.show()
```

3. SETTING THE EDGE COLOR

```python
plt.pie(slice, labels=labels,wedgeprops={'edgecolor':'black'})
plt.show()
```

4. SETTING COLOR TO THE SLICES

```python
color = ['red','blue','green','orange','yellow']
plt.pie(slice,labels=labels,colors=color,wedgeprops={'edgecolor':'black'})
plt.show()
```

5. PLOTTING REAL WORLD DATA

```python
labels = ['JavaScript','HTML/CSS','SQL','Python','Java']
slice = [59219,55466,47544,36443,35917]
plt.pie(slice,labels=labels,colors=color,wedgeprops={'edgecolor':'black'})
plt.show()
```

6. EXPLODING THE PIE CHART

```python
explode = [0,0,0,0.1,0]
plt.pie(slice,labels=labels,colors=color,explode=explode,wedgeprops={'edgecolor':'black'})
plt.show()
```

7. ADDING SHADOW TO THE PIE CHART

```python
plt.pie(slice,labels=labels,colors=color,explode=explode,shadow=True,wedgeprops={'edgecolor':'black'})
plt.show()
```

8. SETTIGN THE STARTING ANGLE

```python
plt.pie(slice,labels=labels,colors=color,explode=explode,startangle=90,wedgeprops={'edgecolor':'black'})
plt.show()
```

9. DISPLAYING PERCENTAGE OF EACH SLICES

```python
plt.pie(slice,labels=labels,colors=color,explode=explode,autopct="%0.1f%%",wedgeprops={'edgecolor':'black'})
plt.show()
```
