PIE CHARTS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

RECOMMENDED TO USE WHEN WE HAVE 5 OR LESS VALUES TO PLOT

1. Plotting the pie chart graph 

```python
slice=[30,40,20,50]
plt.pie(slice)
plt.show()
```

2. ADDING LABELS TO PIE CHART

```python
labels = ['Thirty','Fourty','Twenty','Fifty']
plt.pie(slice, labels = labels)
plt.show()
```

3. SETTING THE EDGE COLOR

```python
plt.pie(slice, labels=labels,wedgeprops={'edgecolor':'magenta'})
plt.show()
```

4. SETTING COLOR TO THE SLICES

```python
color = ['black','pink','cyan','magenta']
plt.pie(slice,labels=labels,colors=color,wedgeprops={'edgecolor':'yellow'})
plt.show()
```
