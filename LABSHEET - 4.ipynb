```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

1. GENERATING 10 RANDOM NUMBERS IN THE RANGE 25 TO 35

```python
ages = [random.randrange(25,35,1) for ages in range(11)]
ages = sorted(ages,reverse=False)
print(ages)
```

2. GENERATING RANDOM NUMBERS BETWEEN 30K TO 45K

```python
devs = [random.randrange(30000,45000,1500) for num in range(11)]
devs = sorted(devs,reverse=True)
print(devs)
```

3. PLOTTING LINE PLOT

```python
plt.plot(ages,devs)
plt.show()
```

4. GENERATING RANDOM NUMBERS BETWEEN 30K TO 45K

```python
devs = [random.randrange(30000,45000,1) for devs in range(11)]
devs = sorted(devs,reverse = False)
print(devs)
```

5. ADDING TITLE, XLABEL, YLABEL

```python
plt.plot(ages,devs)
plt.title("Median Salary in Dollar with Age")
plt.xlabel("Age")
plt.ylabel("Median Salary in Dollar")
plt.show()
```

6. ADDING MORE PLOTS TO THE SAME GRAPH

```python
ages = [random.randrange(25,35,1) for ages in range(11)]
ages = sorted(ages,reverse=False)
devs = [random.randrange(30000,45000,1) for devs in range(11)]
devs = sorted(devs,reverse = False)
py_devs = [random.randrange(50000,70000)for py_devs in range(11)]
py_devs = sorted(py_devs,reverse=False)
js_devs = [random.randrange(80000,90000)for js_devs in range(11)]
js_devs = sorted(js_devs,reverse = False)
plt.xlabel("Age")
plt.ylabel("Median Salary in $")
plt.title("Median Salary in $ with Age")
plt.plot(ages,devs,label = "All Developers")
plt.plot(ages,py_devs,label = "Python Developers")
plt.plot(ages,js_devs,label = "JavaScript Developers")
plt.legend(loc = "lower right")
plt.show()
```

7. SETTING AMRKERS, LINE STYLE AND COLOR

```python
plt.xlabel("Age")
plt.ylabel("Median Salary in $")
plt.title("Median Salary in $ with Age")
plt.plot(ages,devs,color = "#B00B69",linestyle = "--",marker = "D",label = "All Developers")
plt.plot(ages,py_devs,color = "#FF33E9", linestyle = "-.",marker = "*",label = "Python Developers")
plt.plot(ages,js_devs,color = "#FF3355", linestyle = ":",marker = "s",label = "JavaScript Developers")
plt.legend(loc = "lower right")
plt.show()
```

8. CHANGE THE LINE WIDTH

```python
plt.plot(ages,devs,color = "#B00B69",linestyle = "--",marker = "D",linewidth = 2,label = "All Developers")
plt.plot(ages,py_devs,color = "#FF33E9", linestyle = "-.",marker = "*",linewidth = 3,label = "Python Developers")
plt.plot(ages,js_devs,color = "#FF3355", linestyle = ":",marker = "s",linewidth = 4,label = "JavaScript Developers")
```

9. ADDING PADDING TO THE PLOT

```python
plt.xlabel("Age")
plt.ylabel("Median Salary in $")
plt.title("Median Salary in $ with Age")
plt.plot(ages,devs,color = "#B00B69",linestyle = "--",marker = "D",label = "All Developers")
plt.plot(ages,py_devs,color = "#FF33E9", linestyle = "-.",marker = "*",label = "Python Developers")
plt.plot(ages,js_devs,color = "#FF3355", linestyle = ":",marker = "s",label = "JavaScript Developers")
plt.legend(loc = "lower right")
plt.tight_layout() #adds padding
plt.show()
```

10. ADDING GRID TO THE PLOT

```python
plt.xlabel("Age")
plt.ylabel("Median Salary in $")
plt.title("Median Salary in $ with Age")
plt.plot(ages,devs,color = "#B00B69",linestyle = "--",marker = "D",label = "All Developers")
plt.plot(ages,py_devs,color = "#FF33E9", linestyle = "-.",marker = "*",label = "Python Developers")
plt.plot(ages,js_devs,color = "#FF3355", linestyle = ":",marker = "s",label = "JavaScript Developers")
plt.legend(loc = "lower right")
plt.grid(True)
plt.tight_layout()
plt.show()
```
