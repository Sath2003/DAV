```python
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
```

1. GENERATING 10 RANDOM NUMBERS IN THE RANGE 25 TO 35

```python
ages = [random.randrange(25,35,1) for ages in range(11)]
ages = sorted(ages,reverse=False)
print(ages)
```

2. GENERATING RANDOM NUMBERS BETWEEN 30K TO 45K

```python
devs = [random.randrange(30000,45000,1500) for num in range(11)]
devs = sorted(devs,reverse=True)
print(devs)
```

3. PLOTTING LINE PLOT

```python
plt.plot(ages,devs)
plt.show()
```
