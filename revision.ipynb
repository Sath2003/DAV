```python
import pandas as pd 
df = pd.read_csv('/workspaces/DAV/fdata.csv')
df.head()
date = df['Date']
open = df['Open']
high = df['High']
low = df['Low']
close = df['Close']
plt.plot(date,open,label="Open",marker="*",color="green",linestyle='dashed')
plt.plot(date,high,label="High",marker="^",color="blue")
plt.plot(date,low,label="Low",marker="s",color="orange")
plt.plot(date,close,label="Close",marker="o",color="red")
plt.legend()
```

```python
df = pd.read_csv('/workspaces/DAV/data.csv')
df.head()
id = df['Responder_id']
lang = df['LanguagesWorkedWith']
from collections import Counter
lang_counter = Counter()
for res in lang:
    lang_counter.update(res.split(';'))
lang = []
popularity = []
for item in lang_counter.most_common(15):
    lang.append(item[0])
    popularity.append(item[1])
plt.barh(lang,popularity,color="red")
plt.title("Languages and thier Popularity")
plt.ylabel("Languages")
plt.xlabel("Popularity")
```
